In this notebook, I improvised the WordCloud module developed by Andreas Mueller https://peekaboo-vision.blogspot.com/2012/11/a-wordcloud-in-python.html, to allow more artistic creation of word cloud.  My modified class WordArt allows the specification of primary keyword (featured in the center of the word cloud), secondary keywords (scattered around the primary keyword), and a list of background words and symbols.  The core class IntegralOccupancyMap in the original WordCloud module is preserved for plotting the background words only, in order to populate the entire canvas.  Furthermore, my improvised class allows the specification of different font styles and font colors for the primary, secondary, and background words and symbols.  

To make such improvisation, I dived deep into the original WordCloud module and learned to use the pillow library and matplotlib colormap functionality.  In addition, I custom designed several color palettes for harmony, aesthetic, and fun!  This is still work in progress, and I planned to add other functinalities such as support for foreign languages in the future!

```python
import os, random, sys
from random import Random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# For creating word cloud.
from PIL import Image, ImageColor, ImageDraw, ImageFilter, ImageFont, ImageOps
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from operator import itemgetter
from query_integral_image import query_integral_image
```

My modified class is as follows.

```python
# The class for populating words on canvas.
class IntegralOccupancyMap(object):
    def __init__(self, height, width, mask):
        self.height = height
        self.width = width
        if mask is not None:
            # the order of the cumsum's is important for speed ?!
            self.integral = np.cumsum(np.cumsum(255*mask, axis=1), axis=0).astype(np.uint32)
        else:
            self.integral = np.zeros((height, width), dtype=np.uint32)

    def sample_position(self, size_x, size_y, random_state):
        return query_integral_image(self.integral, size_x, size_y, random_state)

    def update(self, img_array, pos_x, pos_y):
        partial_integral = np.cumsum(np.cumsum(img_array[pos_x:, pos_y:], axis=1), axis=0)
        # paste recomputed part into old image
        # if x or y is zero it is a bit annoying
        if pos_x > 0:
            if pos_y > 0:
                partial_integral += (self.integral[pos_x - 1, pos_y:] - self.integral[pos_x - 1, pos_y - 1])
            else:
                partial_integral += self.integral[pos_x - 1, pos_y:]
        if pos_y > 0:
            partial_integral += self.integral[pos_x:, pos_y - 1][:, np.newaxis]
        self.integral[pos_x:, pos_y:] = partial_integral

# Color map.
class colormap_color_func(object):
    """Color func created from matplotlib colormap.

    Parameters
    ----------
    colormap : A string or matplotlib colormap to sample from.

    Example
    -------
    >>> WordCloud(color_func=colormap_color_func("Purples"))

    """
    def __init__(self, colormap):
        import matplotlib.pyplot as plt
        self.colormap = plt.cm.get_cmap(colormap)

    def __call__(self, word, font_size, position, orientation, random_state=None, **kwargs):
        if random_state is None:
            random_state = random.Random()
        r, g, b, _ = np.maximum(0, 255*np.array(self.colormap(random_state.uniform(0, 1))))
        return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

# My new WordArt class.
class WordArt(WordCloud):
    def __init__(self, lists, frequencies, 
                 lists_font_paths=None, font_path=None, symbol_path=None, 
                 lists_color_funcs=None, color_func=None, symbol_color_func=None,
                 lists_colormap=None, colormap=None, symbol_colormap=None,
                 width=1800, height=1800, mask=None, 
                 contour_width=0, contour_color="black", background_color="white", 
                 use_mask_as_bg=False, br=0, mode="RGB", scale=1, 
                 min_font_size=4, max_font_size=200, font_refac=[1], font_step=1, repeat=True, max_words=200, 
                 prefer_horizontal=.9, margin=2, relative_scaling="auto", 
                 min_word_length=0, stopwords=None, normalize_plurals=True, include_numbers=False, regexp=None, 
                 collocations=True, collocation_threshold=30, ranks_only=None, 
                 random_state=None):
        WordCloud.__init__(self, font_path, width, height, margin, ranks_only, prefer_horizontal, mask, scale, 
                           color_func, max_words, min_font_size, stopwords, random_state, background_color, 
                           max_font_size, font_step, mode, relative_scaling, regexp, collocations, colormap, 
                           normalize_plurals, contour_width, contour_color, repeat, include_numbers, min_word_length, 
                           collocation_threshold)
        self.lists_font_paths = lists_font_paths
        self.symbol_path = symbol_path
        self.lists_color_funcs = lists_color_funcs or [colormap_color_func(c) for c in lists_colormap]
        self.symbol_color_func = symbol_color_func or colormap_color_func(symbol_colormap)
        self.font_refac = font_refac
        self.use_mask_as_bg = use_mask_as_bg
        self.br = br
        
    def generate_from_lists(self, lists, frequencies): 
        """Create a word_cloud from a dict of lists.

        Parameters
        ----------
        lists: dict of lists of strings
            A dictionary containing lists of primary and secondary words.
        frequencies : dict from (string, binary) to float
            A dictionary containaining (background words, 0/1 for indication of symbols) and associated frequencies.

        Return values
        -------------
        self

        """
        # Set seed for reproducibility.
        if self.random_state is not None:
            random_state = self.random_state
        else:
            random_state = random.Random()

        # Layout canvas for drawing word cloud.
        if self.mask is None:
            boolean_mask = None
            height = self.height
            width = self.width
        else:
            boolean_mask = self._get_bolean_mask(self.mask)
            height = self.mask.shape[0]
            width = self.mask.shape[1]
        occupancy = IntegralOccupancyMap(height, width, boolean_mask)
        
        # Create image and draw object.
        img_grey = Image.new("L", (width, height))
        draw = ImageDraw.Draw(img_grey)
        img_array = np.asarray(img_grey)
        
        # Initialize the following to store information necessary for reconstructing word cloud later on.
        words, word_types, font_sizes, font_paths, orientations, positions, colors = [], [], [], [], [], [], []

        # Draw primary words given in lists.
        font_size = self.max_font_size #initialize font size as the maximum specified in max_font_size
        for p in range(len(lists)):
            words.extend(lists[str(p)])
            for i, word in enumerate(lists[str(p)]):
                orientation = None #initialize orientation of word as horizontal
                
                # Center the first primary word.
                if (p==0) and (i==0): 
                    font = ImageFont.truetype(self.lists_font_paths[p], font_size)
                    transposed_font = ImageFont.TransposedFont(font, orientation=orientation)
                    w, h = draw.textsize(word, font=transposed_font)
                    y, x = int((width-w)/2), int((height-h)/2.1)
                else:
                    # For all other words, we need to find an unoccupied position to draw.
                    while True:
                        font = ImageFont.truetype(self.lists_font_paths[p], font_size)
                        transposed_font = ImageFont.TransposedFont(font, orientation=orientation)
                        box_size = draw.textsize(word, font=transposed_font)
                        result = occupancy.sample_position(box_size[1]+self.margin, box_size[0]+self.margin, random_state)
                        if result is not None:
                            break #found a position
                        else:
                            font_size -= self.font_step #reduce font size and keep seaching for position to draw
                    # Draw the word.
                    x, y = np.array(result) + self.margin // 2
                draw.text((y, x), word, fill="white", font=transposed_font)
                # Update information.
                word_types.append(("l", p))
                font_sizes.append(font_size)
                font_paths.append(self.lists_font_paths[p])
                orientations.append(orientation)
                positions.append((x, y))
                colors.append(self.lists_color_funcs[p](word, font_size=font_size, font_path=self.lists_font_paths[p], 
                                                        orientation=orientation, position=(x, y), 
                                                        random_state=random_state))                                
                # Recompute integral image.
                if self.mask is None:
                    img_array = np.asarray(img_grey)
                else:
                    img_array = np.asarray(img_grey) + boolean_mask
                occupancy.update(img_array, x, y)
            # Reduce font size by a factor.
            f = self.font_refac[i] if len(self.font_refac)==len(lists) else self.font_refac[0]
            font_size = int(font_size * f)
                
        # Draw background words and symbols given in the frequencies dictionary.  
        # First, ensure that frequencies are sorted and normalized.
        frequencies = sorted(frequencies.items(), key=itemgetter(1), reverse=True) #this is a list of (key, freq) tuple
        # Retain up to max_words.
        frequencies = frequencies[:self.max_words]
        # Set the highest frequency to 1.
        max_frequency = float(frequencies[0][1])
        frequencies = [(word, freq / max_frequency) for word, freq in frequencies]
        # If repeat=True and we do not have enough words yet, pad frequencies with repeating words.
        if self.repeat and len(frequencies) < self.max_words:
            times_extend = int(np.ceil(self.max_words/len(frequencies))) - 1
            # Get the lowest frequency.
            frequencies_org = list(frequencies)
            downweight = frequencies[-1][1]
            for i in range(times_extend):
                frequencies.extend([(word, freq*downweight**(i + 1)) for word, freq in frequencies_org])      
        # Now, draw background words and symbols based on frequencies.
        last_freq = 1
        for (word, s), freq in frequencies:
            # Set font_path and color_func dependeing on whether the word is a symbol or a regular word.
            if s==1:
                font_path = self.symbol_path
                color_func = self.symbol_color_func
            else:
                font_path = self.font_path
                color_func = self.color_func
            # Set font size and orientation.
            if freq == 0:
                continue
            # select the font size
            rs = self.relative_scaling
            if rs != 0:
                font_size = int(round((rs * (freq / float(last_freq)) + (1 - rs)) * font_size))
            # Optinally rotate words.
            if random_state.random() < self.prefer_horizontal:
                orientation = None
            else:
                orientation = Image.ROTATE_90
            tried_other_orientation = False
            # Find a position to draw.
            while True:
                font = ImageFont.truetype(font_path, font_size)
                # Transpose font based on orientation.
                transposed_font = ImageFont.TransposedFont(font, orientation=orientation)
                # Get size of the resulting word.
                box_size = draw.textsize(word, font=transposed_font)
                # Find possible position to draw the word or symbol.
                result = occupancy.sample_position(box_size[1]+self.margin, box_size[0]+self.margin, random_state)
                if result is not None or font_size < self.min_font_size:
                    break #either we found a poition to draw or font size went too small
                # If we didn't find a place, try rotation.  Otherwise, use smaller font size.
                if not tried_other_orientation and self.prefer_horizontal < 1:
                    orientation = (Image.ROTATE_90 if orientation is None else Image.ROTATE_90)
                    tried_other_orientation = True
                else:
                    font_size -= self.font_step
                    orientation = None
            # Stop drawing if font size went too small.
            if font_size < self.min_font_size:
                break
            # Draw the word.
            x, y = np.array(result) + self.margin // 2
            draw.text((y, x), word, fill="white", font=transposed_font)
            # Update information.
            word_types.append("w" if s==0 else "s")
            words.append(word)
            font_sizes.append(font_size)
            font_paths.append(font_path)
            orientations.append(orientation)
            positions.append((x, y))
            colors.append(color_func(word, font_size=font_size, font_path=font_path,
                                     orientation=orientation, position=(x, y), random_state=random_state))
            # Recompute integral image.
            if self.mask is None:
                img_array = np.asarray(img_grey)
            else:
                img_array = np.asarray(img_grey) + boolean_mask
            occupancy.update(img_array, x, y)
            last_freq = freq
        
        # Return stored information for drawing word cloud.
        self.layout_ = list(zip(words, word_types, font_sizes, font_paths, positions, orientations, colors))
        return self
    
    def recolor(self, lists_color_funcs=None, color_func=None, symbol_color_func=None,
                lists_colormap=None, colormap=None, symbol_colormap=None, random_state=None):
        """Recolor existing layout.
        Applying a new coloring is much faster than generating the whole wordcloud.
        
        Parameters
        ----------
        color_func : function or None, default=None
            Function to generate new color from word count, font size, position
            and orientation.  If None, self.color_func is used.
        colormap : string or matplotlib colormap, default=None
            Use this colormap to generate new colors. Ignored if color_func
            is specified. If None, self.color_func (or self.color_map) is used.
        random_state : RandomState, int, or None, default=None
            If not None, a fixed random state is used. If an int is given, this
            is used as seed for a random.Random state.
            
        Returns
        -------
        self
        """
        if isinstance(random_state, int):
            random_state = Random(random_state)
        self._check_generated()

        # Set new color functions.
        lists_color_funcs = lists_color_funcs or [colormap_color_func(c) for c in lists_colormap]
        color_func = color_func or colormap_color_func(colormap)
        symbol_color_func = symbol_color_func or colormap_color_func(symbol_colormap)
        # Reset self.layout_ with new color functions.
        def map_color(word_type):
            if isinstance(word_type, tuple):
                return lists_color_funcs[word_type[1]]
            elif word_type=="w":
                return color_func
            else:
                return symbol_color_func
            
        self.layout_ = [(word, word_type, font_size, font_path, position, orientation, 
                         map_color(word_type)(word=word, font_size=font_size, 
                                              position=position, orientation=orientation,
                                              random_state=random_state))
                        for word, word_type, font_size, font_path, position, orientation, _ in self.layout_]
        return self
    
    def to_image(self):
        """Draw word cloud based on information stored in self.layout_.

        Returns
        -------
        word cloud image

        """
        self._check_generated()
        if self.mask is not None:
            height = self.mask.shape[0]
            width = self.mask.shape[1]
        else:
            height = self.height
            width = self.width
            
        if self.use_mask_as_bg:
            # Black mask only:
            #img = Image.fromarray(self.mask).resize((int(width*self.scale), int(height*self.scale)))
            #if self.br>0:
            #    img = img.filter(ImageFilter.GaussianBlur(self.br))
                
            #img = Image.new(self.mode, (int(width*self.scale), int(height*self.scale)), self.background_color)
            #mask = Image.fromarray(self.mask).convert("L").resize((int(width*self.scale), int(height*self.scale)))
            #if self.br>0:
            #    mask = mask.filter(ImageFilter.GaussianBlur(self.br))
            #img.paste(mask, (0,0), mask)
            r, g, b = color_dict[self.background_color]
            mask_array = np.array(self.mask)
            pos_to_update = np.any(mask_array!=255, axis=-1)
            mask_array[pos_to_update] = (r, g, b)
            img = Image.fromarray(mask_array).resize((int(width*self.scale), int(height*self.scale)))
            if self.br>0:
                img = img.filter(ImageFilter.GaussianBlur(self.br))
        else:
            img = Image.new(self.mode, (int(width*self.scale), int(height*self.scale)), self.background_color)
        
        draw = ImageDraw.Draw(img)
        for word, word_type, font_size, font_path, position, orientation, color in self.layout_:
            font = ImageFont.truetype(font_path, int(font_size*self.scale))
            transposed_font = ImageFont.TransposedFont(font, orientation=orientation)
            pos = (int(position[1]*self.scale), int(position[0]*self.scale))
            draw.text(pos, word, fill=color, font=transposed_font)

        return self._draw_contour(img=img)
```

The following functions are for showing, resizing or saving the generated word clouds.

```python
def show_image(wordart):
    plt.figure(figsize=(10, 10), facecolor=None) 
    plt.axis("off")
    plt.tight_layout(pad=0) 
    plt.imshow(wordart, interpolation="bilinear")
    plt.show()

def resize_image(image, factor, new_filename):
    new_size = [int(factor * s) for s in image.size]
    image_resize = image.resize(new_size, Image.ANTIALIAS)
    image_resize.save(new_filename, dpi=(300, 300))
    
# https://pillow.readthedocs.io/en/5.1.x/handbook/tutorial.html
# https://github.com/amueller/word_cloud/blob/master/wordcloud/wordcloud.py
# https://www.pngfuel.com/
def set_image_dpi_resize(image):
    """
    Rescaling image to 300dpi while resizing
    Parameters: An image
    Return values: A rescaled image
    """
    height, width = image.size
    factor = min(1, float(1024.0 / height))
    new_size = int(factor * height), int(factor * width)
    image_resize = image.resize(new_size, Image.ANTIALIAS)
    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix='1.jpg')
    temp_filename = temp_file.name
    image_resize.save(temp_filename, dpi=(300, 300))
    return temp_filename

def set_image_dpi(image):
    """
    Rescaling image to 300dpi without resizing
    Parameters: An image
    Return values: A rescaled image
    """
    temp_file = tempfile.NamedTemporaryFile(delete=False, suffix='.jpg')
    temp_filename = temp_file.name
    image.save(temp_filename, dpi=(300, 300))
    return temp_filename
```

The following function generates word cloud from text (instead of a dict of words and frequencies as in the word_cloud module).  The text will first go though natural language processing, and will be converted into a dictionary of words and frequencies for word cloud generation.

```python
def createWordCloud(text, font_path=None, color_func=None, colormap=None, 
                    width=2200, height=2800, mask_path=None, contour_width=0, contour_color="black", 
                    background_color="white", mode="RGB", scale=1, 
                    min_font_size=4, max_font_size=None, font_step=1, repeat=True, max_words=200, 
                    prefer_horizontal=.9, margin=2, relative_scaling=0, 
                    save_path=None, random_state=0):
    # If mask_path is not none, set mask.
    if mask_path is not None:
        mask = np.array(Image.open(mask_path))
        print("mask shape (h, w, c) = " + str(mask.shape))
    else:
        mask = None
        print("image shape (h, w) = ({}, {})".format(height, width))

    # Initialize word cloud object.
    wc = WordCloud(font_path=font_path, color_func=color_func, colormap=colormap, 
                   width=width, height=height, mask=mask, contour_width=contour_width, contour_color=contour_color, 
                   background_color=background_color, mode=mode, scale=scale,            
                   min_font_size=min_font_size, max_font_size=max_font_size, font_step=font_step, 
                   repeat=repeat, max_words=max_words, 
                   prefer_horizontal=prefer_horizontal, margin=margin, relative_scaling=relative_scaling,
                   random_state=random_state)
    
    # Generate word cloud.
    wc.generate(text)
    # Plot word cloud.
    plt.figure(figsize=(10, 10), facecolor=None) 
    plt.axis("off")
    plt.tight_layout(pad=0) 
    plt.imshow(wc, interpolation="bilinear")
    plt.show()
    
    # If save_path is specified, save word art.
    if save_path is not None:
        image = wc.to_image()
        image.save(save_path, dpi=(300,300))   
    return wc
    
# generate_from_frequencies
```

The following function generates word cloud that allows specification of primary and second words (through lists) and background words and symbols  (through frequencies).

```python
def createWordArt(lists, frequencies, 
                  lists_font_paths=None, font_path=None, symbol_path=None,
                  lists_color_funcs=None, color_func=None, symbol_color_func=None,
                  lists_colormap=None, colormap=None, symbol_colormap=None,
                  width=2200, height=2800, mask_path=None, contour_width=0, contour_color="black", 
                  background_color="white", use_mask_as_bg=False, br=0, mode="RGB", scale=1, 
                  min_font_size=4, max_font_size=None, font_refac=[2/3], font_step=1, repeat=True, max_words=150, 
                  prefer_horizontal=.9, margin=2, relative_scaling=0, 
                  save_path=None, random_state=0):
    # If mask_path is not none, set mask.
    if mask_path is not None:
        mask = np.array(Image.open(mask_path))
        print("mask shape (h, w, c) = " + str(mask.shape))
    else:
        mask = None
        print("image shape (h, w) = ({}, {})".format(height, width))
    
    # Initialize word art object.
    wa = WordArt(lists=lists, frequencies=frequencies, 
                 lists_font_paths=lists_font_paths, font_path=font_path, symbol_path=symbol_path,
                 lists_color_funcs=lists_color_funcs, color_func=color_func, symbol_color_func=symbol_color_func,
                 lists_colormap=lists_colormap, colormap=colormap, symbol_colormap=symbol_colormap,
                 width=width, height=height, mask=mask, contour_width=contour_width, contour_color=contour_color, 
                 background_color=background_color, use_mask_as_bg=use_mask_as_bg, br=br, mode=mode, scale=scale,            
                 min_font_size=min_font_size, max_font_size=max_font_size, font_refac=font_refac, font_step=font_step, 
                 repeat=repeat, max_words=max_words, 
                 prefer_horizontal=prefer_horizontal, margin=margin, relative_scaling=relative_scaling,
                 random_state=random_state)
    
    # Generate word art.
    wa.generate_from_lists(lists=lists, frequencies=frequencies)
    # Plot word art.
    plt.figure(figsize=(10, 10), facecolor=None) 
    plt.axis("off")
    plt.tight_layout(pad=0) 
    plt.imshow(wa, interpolation="bilinear")
    plt.show()
    
    # If save_path is specified, save word art.
    if save_path is not None:
        image = wa.to_image()
        image.save(save_path, dpi=(300,300))   
    return wa
```

In the following cell, I defined a set of aesthetic elements such as mask, font styles (creativemarket https://creativemarket.com/fonts sells many beautiful commercial fonts) and color palettes.

```python
color_dict = {"black": (0, 0, 0), "white": (255, 255, 255), "gray": (169, 169, 169),
                  "gold": (210, 175, 54), "silver": (170, 169, 173), "pink gold": (183, 110, 120), 
                  "gun metal": (42, 52, 57), "dark gun metal": (31, 38, 42), "smoky black": (24, 2, 0), 
                  "black coral": (91, 103, 109), "chinese gold": (204, 153, 0),
                  "red": (255, 0, 0), "green": (0, 255, 0), "blue": (0, 0, 255),
                  "royal green": (18, 99, 6), "slimy green": (38, 141, 25), "midnight green": (0, 86, 96),
                  "teal": (0, 128, 129), "purple": (125, 0, 130),
                  "royal blue": (57, 82, 217), "indigo dye": (22, 39, 146), "navy": (0, 1, 131), 
                  "rose garnet": (153, 0, 68),  "dogwood rose": (216, 27, 102),
                  "barbie pink": (224, 33, 139), "mexican pink": (228, 0, 124), "philippine pink": (250, 26, 142),
                  "pale pink": (255,231,248), "light pink": (255,207,238), "light purple": (200,187,255),
                  "light blue": (191,240,255), "pale blue": (223,248,255)}

# Mask.
MASK_DIR = "./"
mask_dict = {"heart": "heart.jpg"}
def get_mask_path(mask_name):
    return os.path.join(MASK_DIR, mask_dict[mask_name])

# Font.
"""
/System/Library/Fonts
/Library/Fonts
"""
XL_FONT_DIR = "./" #excluded due to commercial font licensing restriction
SYS_FONT_DIR = "/System/Library/Fonts"
FONT_DIR = "/Library/Fonts"
XL_font_dict = {"BlackChancery": "blackchancery/BLKCHCRY.TTF", 
                "AmalfiCoast":"Amalfi-Coast-Chic-Calligraphy./Amalfi Coast.otf",
                "HelloHoney": "hello-honey-a-chic-script-font/hello honey.otf", 
                "Symbola": "symbola/Symbola.ttf"}
sys_font_dict = {"Menlo": "Menlo.ttc"}
font_dict = {"Futura": "Futura.ttc", "TimesNewRoman": "Times New Roman.ttf"}
def get_font_path(font_name):
    if font_name in XL_font_dict:
        p = os.path.join(XL_FONT_DIR, XL_font_dict[font_name])
    elif font_name in sys_font_dict:
        p = os.path.join(SYS_FONT_DIR, sys_font_dict[font_name])
    else:
        p = os.path.join(FONT_DIR, font_dict[font_name])
    return p


# Color.
"""
HSL: H=hue, 0 (red), 120 (green), 240 (blue)
     S=saturation, 0 (gray) - 100 (full color)
     L=lightness, 0 (black) - 100 (white)
     A=alpha, 0 (fully transpararent) - 1 (not transparent)
HSV: V=value=B=brightness, 0 (black) - 100 (no black)

color palette:
https://www.schemecolor.com
https://colorswall.com/palettes

matplotlib colormap:
https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
https://matplotlib.org/gallery/color/colormap_reference.html
print(plt.colormaps())
"""

# Single color.
def get_color_func(color_name):
    color_dict = {"black": (0, 0, 0), "white": (255, 255, 255), "gray": (169, 169, 169),
                  "gold": (210, 175, 54), "silver": (170, 169, 173), "pink gold": (183, 110, 120), 
                  "gun metal": (42, 52, 57), "dark gun metal": (31, 38, 42), "smoky black": (24, 2, 0), 
                  "black coral": (91, 103, 109), "chinese gold": (204, 153, 0),
                  "red": (255, 0, 0), "green": (0, 255, 0), "blue": (0, 0, 255),
                  "royal green": (18, 99, 6), "slimy green": (38, 141, 25), "midnight green": (0, 86, 96),
                  "teal": (0, 128, 129), "purple": (125, 0, 130),
                  "royal blue": (57, 82, 217), "indigo dye": (22, 39, 146), "navy": (0, 1, 131), 
                  "rose garnet": (153, 0, 68),  "dogwood rose": (216, 27, 102),
                  "barbie pink": (224, 33, 139), "mexican pink": (228, 0, 124), "philippine pink": (250, 26, 142),
                  "pale pink": (255,231,248), "light pink": (255,207,238), "light purple": (200,187,255),
                  "light blue": (191,240,255), "pale blue": (223,248,255)}
    r, g, b = color_dict[color_name]
    
    def single_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None, **kwargs): 
        return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)
    
    return single_color_func

# Random color.
def random_color_func(word=None, font_size=None, position=None, orientation=None, font_path=None, random_state=None, **kwargs):
    """Random hue color generation.

    Default coloring method. This just picks a random hue with value 80% and
    lumination 50%.

    Parameters
    ----------
    word, font_size, position, orientation  : ignored.

    random_state : random.Random object or None, (default=None)
        If a random object is given, this is used for generating random
        numbers.

    """
    if random_state is None:
        random_state = Random()
    return "hsl(%d, 80%%, 50%%)" % random_state.randint(0, 255)

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

def darkgrey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(0, 40)

def red_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<0.5:
        g = "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
    else:
        g = "hsl(320, 100%%, %d%%)" % random.randint(60, 100)
    return g

def allpink_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<1/4:
        r, g, b = 252, 90, 141
    elif (x>=1/4) & (x<1/2):
        r, g, b = 252, 191, 208
    elif (x>=1/2) & (x<3/4):
        r, g, b = 255, 231, 248
    else: 
        r, g, b = 255,207,238
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

def babypink_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<2/4:
        r, g, b = 252, 90, 141
    elif (x>=2/4) & (x<3/4):
        r, g, b = 252, 191, 208
    else:
        r, g, b = 173, 175, 178
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

def babyblue_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<1/3:
        r, g, b = 51, 125, 232 #bleu de france
    elif (x>=1/3) & (x<2/3):
        r, g, b = 136, 206, 241 #baby blue
    else:
        r, g, b = 173, 175, 178
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

def green_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<1/3:
        r, g, b = 38, 141, 25 #slimy green
    elif (x>=1/3) & (x<2/3):
        r, g, b = 124, 212, 91 #mantis
    else:
        r, g, b = 173, 175, 178
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

def rose_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<1/6:
        r, g, b = 153, 0, 68 #rose garnet (*)
    elif (x>=1/6) & (x<1/3):
        r, g, b = 193, 28, 85 #rose red (*)
    elif (x>=1/3) & (x<1/2):
        r, g, b = 253, 85, 113 #fiery rose
    elif (x>=1/2) & (x<2/3):
        r, g, b = 139, 19, 56 #claret
    elif (x>=2/3) & (x<5/6):
        r, g, b = 249, 76, 140 #french rose
    else:
        r, g, b = 216, 27, 102 #dogwood rose (*)
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

def teal_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<1/6:
        r, g, b = 0, 86, 96 #midnight green (*)
    elif (x>=1/6) & (x<1/3):
        r, g, b = 0, 116, 117 #skobeloff
    elif (x>=1/3) & (x<1/2):
        r, g, b = 0, 128, 129 #teal (*)
    elif (x>=1/2) & (x<2/3):
        r, g, b = 34, 147, 137 #illuminating emerald
    elif (x>=2/3) & (x<5/6):
        r, g, b = 52, 167, 152 #zomp
    else:
        r, g, b = 87, 195, 173 #verdigris
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

def maleficent_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    x = Random().random()
    if x<1/6:
        r, g, b = 48,26,57 #dark purple (as background)
    elif (x>=1/6) & (x<1/3):
        r, g, b = 76,22,102 #purple
    elif (x>=1/3) & (x<1/2):
        r, g, b = 114,33,111 #red purple
    elif (x>=1/2) & (x<2/3):
        r, g, b = 0, 255, 0 #lime green
    elif (x>=2/3) & (x<5/6):
        r, g, b = 43,170,58 #green 
    else:
        r, g, b = 43,100,58 #dark green
    return "rgb({:.0f}, {:.0f}, {:.0f})".format(r, g, b)

#color_func = ImageColorGenerator(mask)
```
